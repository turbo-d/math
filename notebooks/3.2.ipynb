```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def xy_cart_plane(size):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.set_aspect('equal')
    
    ax.set_xlim([-(size+1), size+1])
    ax.set_ylim([-(size+1), size+1])
    
    ax.spines.left.set_position('zero')
    ax.spines.bottom.set_position('zero')
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
    
    ticks = np.arange(-size, size+1)
    ax.set_xticks(ticks[ticks != 0])
    ax.set_yticks(ticks[ticks != 0])
    
    ax.grid(color='grey', linestyle='dotted')
    
    ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
    ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
    
    ax.set_xlabel('x', size=14, labelpad=-23, x=1.04)
    ax.set_ylabel('y', size=14, labelpad=-23, y=1.04, rotation='horizontal')

    return ax
```

```python
# 39
ax = xy_cart_plane(5)

x1s = np.arange(-5, 1, .1)
y1s = 2*x1s - 1

x2s = np.arange(1, 6, .1)
y2s = 1 + x2s

ax.plot(x1s, y1s)
ax.plot(x2s, y2s)
```

```python
# 41
ax = xy_cart_plane(5)

x1s = np.arange(-5, 0, .1)
y1s = np.full(len(x1s), 3)

x2s = np.arange(0, 6, .1)
y2s = np.sqrt(x2s)

ax.plot(x1s, y1s)
ax.plot(x2s, y2s)
```
