```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def xy_cart_plane(size):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.set_aspect('equal')
    
    ax.set_xlim([-(size+1), size+1])
    ax.set_ylim([-(size+1), size+1])
    
    ax.spines.left.set_position('zero')
    ax.spines.bottom.set_position('zero')
    ax.spines.right.set_visible(False)
    ax.spines.top.set_visible(False)
    
    ticks = np.arange(-size, size+1)
    ax.set_xticks(ticks[ticks != 0])
    ax.set_yticks(ticks[ticks != 0])
    
    ax.grid(color='grey', linestyle='dotted')
    
    ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
    ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)
    
    ax.set_xlabel('x', size=14, labelpad=-23, x=1.04)
    ax.set_ylabel('y', size=14, labelpad=-23, y=1.04, rotation='horizontal')

    return ax
```

```python
# 25
ax = xy_cart_plane(5)

xs = np.arange(-5, 6, .1)
ys = 2**xs - 3

ax.plot(xs, ys)
```

```python
ax = xy_cart_plane(5)

# f(x) = x**2
xs = np.arange(-5, 6, .1)
ys = (xs)**2
ax.plot(xs, ys)

# g(x) = f(x+1)
# g(x) = (x+1)**2
y1s = (xs + 1)**2
ax.plot(xs, y1s)

# h(x) = 4*g(x)
# h(x) = 4*(x+1)**2
y2s = 4*(xs + 1)**2
ax.plot(xs, y2s)

# i(x) = h(x) - 5
# i(x) = 4*(x+1)**2 - 5
y2s = 4*(xs + 1)**2 - 5
ax.plot(xs, y2s)
```

```python
ax = xy_cart_plane(5)

# f(x) = x**2
xs = np.arange(-5, 6, .1)
ys = (xs)**2
ax.plot(xs, ys)

# g(x) = f(2*x)
# g(x) = (2*x)**2
y1s = (2*xs)**2
ax.plot(xs, y1s)

# h(x) = g(x+1) = f(2*(x+1)) = f(2*x + 2)
# h(x) = (2*(x + 1))**2
y1s = (2*(xs + 1))**2
ax.plot(xs, y1s)

# i(x) = h(x) - 5
# i(x) = (2*(x + 1))**2 - 5
y1s = (2*(xs + 1))**2 - 5
ax.plot(xs, y1s)
```
